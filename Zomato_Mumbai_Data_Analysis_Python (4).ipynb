**Q7) Which areas have a large number of Chinese Restaurant Market?**


```python
chinese_df = zomato_df[
    zomato_df['CUSINE_CATEGORY'].str.contains('Chinese', case=False, na=False)
]

chinese_df
```

What This Code Does (Simple)

1️⃣ Groups Chinese restaurants by Region
2️⃣ Counts number of restaurants (NAME → count)
3️⃣ Calculates average price (PRICE → mean)
4️⃣ Renames columns properly
5️⃣ Sorts by highest number of restaurants
6️⃣ Shows top 25 regions

```python
chinese_rest_df = (
    chinese_df
        .groupby('REGION')
        .agg({
            'NAME': 'count',
            'PRICE': 'mean'
        })
        .rename(columns={
            'NAME': 'COUNT OF RESTAURANTS',
            'PRICE': 'AVERAGE PRICE'
        })
        .reset_index()
)

chinese_rest_df = chinese_rest_df.sort_values(
    by='COUNT OF RESTAURANTS',
    ascending=False
).head(25)

chinese_rest_df.head()
```

**No. of Chinese Restaurant by Places**


```python
import plotly.express as px

fig = px.bar(
    chinese_rest_df,
    x='REGION',
    y='COUNT OF RESTAURANTS',
    color='AVERAGE PRICE',
    title='Number of Chinese Restaurants by Region',
    labels={
        'REGION': 'Region',
        'COUNT OF RESTAURANTS': 'Number of Restaurants',
        'AVERAGE PRICE': 'Average Price'
    }
)

fig.update_layout(title_x=0.5)

fig.show()
```

**Q10) Find the list of Affordable Restaurants?**
#The criteria for Affordable Restaurants would be:-
1) Low Price
2) High Rated

First step will be to find the restaurants with average cost 1/4th the average cost of most expensive restaurant in our dataframe.

Let me explain:-The most expensive restaurant has an average meal cost= 6000. We'll try to stay economical and only pick
the restaurants that are 1/4th of 6000.


```python
max_price = zomato_df['PRICE'].max()
one_fourth_price = max_price/4
one_fourth_price
```

# Finding list of restaurants that have price less than and equal to 1/4th of the max price i.e Finding Cheap Restauran

```python
# Select required columns
aff_rest_df = zomato_df[
    ['NAME', 'PRICE', 'CUSINE_CATEGORY', 'REGION', 'CUSINE TYPE']
]

# Filter restaurants with price <= 1250
aff_rest_df = aff_rest_df[aff_rest_df['PRICE'] <= 1250]

# Sort by price (lowest first)
aff_rest_df.sort_values(by='PRICE', inplace=True)

aff_rest_df
```

# Finding the highest rated list of restaurants

```python
highrate_rest_df = zomato_df[['NAME', 'PRICE', 'CUSINE_CATEGORY', 'REGION', 'CUSINE TYPE','RATING']]
highrate_rest_df = highrate_rest_df[highrate_rest_df['RATING'] >= 4.5]
highrate_rest_df.sort_values(by='PRICE', inplace=True)
highrate_rest_df
```

**Now, we'll merge the aff_rest_df with highrate_rest_df to obtain the intersection i.e the list of Affordable Restaurants !**

# Affordable Restaurants with low price and high rating


```python
# Merge affordable restaurants with high-rated restaurants
highrate_aff_df = pd.merge(
    aff_rest_df,
    highrate_rest_df,
    how='inner',
    on=['NAME', 'REGION']
)

# Select required columns
highrate_aff_df = highrate_aff_df[
    ['NAME', 'PRICE_x', 'CUSINE_CATEGORY_x', 'REGION', 'CUSINE TYPE_x']
]

# Rename columns properly
highrate_aff_df.rename(
    columns={
        'PRICE_x': 'PRICE',
        'CUSINE_CATEGORY_x': 'CUSINE_CATEGORY',
        'CUSINE TYPE_x': 'CUSINE TYPE'
    },
    inplace=True
)

highrate_aff_df
```
