```python
import pandas as pd
import plotly.express as px

dataset=pd.read_csv("Zomato_Mumbai_Dataset.csv", delimiter='|')
```

```python
dataset.head()
```

**(1) Restaurants by Region (Top 20)**

```python
region_count = dataset['REGION'].value_counts().head(20).reset_index()
region_count.columns = ['Region', 'Restaurant Count']

fig = px.bar(
    region_count,
    x='Region',
    y='Restaurant Count',
    title='Top 20 Regions with Most Restaurants',
)

fig.show()
```

**(2) Cuisine Popularity Chart**

```python
cuisine = dataset['CUSINE_CATEGORY'].str.split(', ').explode()
cuisine_count = cuisine.value_counts().head(15).reset_index()
cuisine_count.columns = ['Cuisine', 'Count']

fig = px.bar(
    cuisine_count,
    x='Cuisine',
    y='Count',
    title='Top 15 Most Popular Cuisines',
)

fig.show()
```

**(3) Rating Distribution**

```python
fig = px.histogram(
    dataset,
    x='RATING',
    nbins=20,
    title='Restaurant Rating Distribution'
)

fig.show()
```

**(4) Price Distribution**

```python
fig = px.histogram(
    dataset,
    x='PRICE',
    nbins=40,
    title='Restaurant Price Distribution'
)

fig.show()
```

**(5) Rating vs Votes (Popularity)**

```python
fig = px.scatter(
    dataset,
    x='VOTES',
    y='RATING',
    title='Rating vs Votes',
    hover_data=['NAME']
)

fig.show()
```

**(6) Average Price by Region**

```python
dataset['PRICE'] = pd.to_numeric(dataset['PRICE'], errors='coerce')
price_region = dataset.groupby('REGION')['PRICE'].mean().sort_values(ascending=False).head(15)

fig = px.bar(
    price_region,
    title='Most Expensive Regions',
)

fig.show()
```

**(7) Rating Type Distribution**

```python
rating_type = dataset['RATING_TYPE'].value_counts().reset_index()
rating_type.columns = ['Rating Type','Count']

fig = px.pie(
    rating_type,
    names='Rating Type',
    values='Count',
    title='Rating Type Distribution'
)

fig.show()
```

**(8) Top 20 Restaurants by Votes**

```python
top_votes = dataset.sort_values(by='VOTES', ascending=False).head(20)

fig = px.bar(
    top_votes,
    x='NAME',
    y='VOTES',
    title='Top 20 Most Popular Restaurants'
)

fig.show()
```

**(9) Top Rated Restaurants**

```python
dataset['RATING'] = pd.to_numeric(dataset['RATING'], errors='coerce')
top_rating = dataset[dataset['RATING']>=4.5].sort_values(by='RATING', ascending=False).head(20)

fig = px.bar(
    top_rating,
    x='NAME',
    y='RATING',
    title='Top Rated Restaurants'
)

fig.show()
```

**(10) Region vs Average Rating**

```python
rating_region = dataset.groupby('REGION')['RATING'].mean().sort_values(ascending=False).head(15)

fig = px.bar(
    rating_region,
    title='Regions with Highest Average Ratings'
)

fig.show()
```

**(11) Price vs Rating Relationship**

```python
fig = px.scatter(
    dataset,
    x='PRICE',
    y='RATING',
    title='Price vs Rating'
)

fig.show()
```

**(12) Cuisine vs Average Rating**

```python
cuisine_rating = dataset.groupby('CUSINE_CATEGORY')['RATING'].mean().sort_values(ascending=False).head(15)

fig = px.bar(
    cuisine_rating,
    title='Cuisine vs Average Rating'
)

fig.show()
```

**(13) Restaurants per Page (Scraping Analysis)**

```python
page_count = dataset['PAGE NO'].value_counts().sort_index()

fig = px.line(
    page_count,
    title='Restaurants Scraped per Page'
)

fig.show()
```

**(14) Top Regions by Votes**

```python
votes_region = dataset.groupby('REGION')['VOTES'].sum().sort_values(ascending=False).head(15)

fig = px.bar(
    votes_region,
    title='Regions with Most Customer Votes'
)

fig.show()
```

**(15)Cuisine vs Price**

```python
cuisine_price = dataset.groupby('CUSINE_CATEGORY')['PRICE'].mean().sort_values(ascending=False).head(15)

fig = px.bar(
    cuisine_price,
    title='Most Expensive Cuisine Types'
)

fig.show()
```
