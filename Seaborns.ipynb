```python
import numpy as np
import pandas as pd

class Descriptive:
    def __init__(self):
        pass

    def segreQuanQual(self, dataset):
        """Separates quantitative and qualitative column names from a dataset."""
        quantitative_cols = []
        qualitative_cols = []
        for column in dataset.columns:
            if dataset[column].dtype in ['int64', 'float64']:
                quantitative_cols.append(column)
            else:
                qualitative_cols.append(column)
        return quantitative_cols, qualitative_cols

    def descriptive_Analysis(self, dataset, quantative_cols):
        """Performs descriptive analysis on quantitative columns."""
        # Placeholder implementation: Returns basic descriptive stats
        return dataset[quantative_cols].describe()

    def outliercolumn(self, quantative_cols, des_Data):
        """Identifies outliers in quantitative columns based on descriptive data."""
        lesser = {}
        greater = {}
        for column in quantative_cols:
            iqr = des_Data.loc['75%', column] - des_Data.loc['25%', column]
            lower_bound = des_Data.loc['25%', column] - 1.5 * iqr
            upper_bound = des_Data.loc['75%', column] + 1.5 * iqr

            # For demonstration, we'll just store the bounds
            # In a real scenario, you'd find actual outlier values
            lesser[column] = lower_bound
            greater[column] = upper_bound
        return lesser, greater

    def changeoutlier(self, dataset, des_Data, lesser, greater):
        """Replaces outliers in the dataset. This is a placeholder.
        A common approach is to replace outliers with the upper/lower bound or NaN.
        """
        # Placeholder implementation: for simplicity, we'll return the original dataset
        # In a real scenario, you would implement outlier capping or removal.
        print("Outlier replacement logic goes here. Returning original dataset for now.")
        # Example of capping outliers:
        # for column in lesser.keys():
        #     dataset.loc[dataset[column] < lesser[column], column] = lesser[column]
        #     dataset.loc[dataset[column] > greater[column], column] = greater[column]
        return dataset

obj=Descriptive()
```
